1. For a group of 7 people, find the probability that all 4 seasons (winter, spring, summer, fall)
occur at least once each among their birthdays, assuming that all seasons are equally likely.


sol : This is based on inclusion and exclusion concept Seven people are there with 4 choice of season so possibility = 4^7 These are the total outcomes Now outcomes with three distinct season = 4C33^7 but in this the outcomnes with two seasons are double counted Now outcomes with two distinct season = 4C22^7 is subtracted out so outcomes wuth one seasons = 4C11^7 Favourable outcomes = 4^7-(4C33^7-4C22^7+4C11^7 )= 8400 Total Outcomes = 4^7

Required probability = 8400/4^7 =0.51269
